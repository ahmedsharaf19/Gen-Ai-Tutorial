```python
# مثال كامل كدا على الموضوع هيمشي ازاى بعدين من اول ما نقرا الداتا ونقسمها لحد ما نخزنها

# 1 - Data Loader
from langchain_community.document_loaders import TextLoader
loader = TextLoader('speech.txt')
docs = loader.load()

# 2 - Data Splitting
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
final_documents = splitter.split_documents(docs)

# 3 - Data Embedding
from langchain_community.embeddings import OllamaEmbeddings
embeddings = OllamaEmbeddings(model = 'mxbai-embed-large')

# 4 - Vector StoreDB
from langchain_community.vectorstores import Chroma
db = Chroma.from_documents(final_documents, embeddings) # (documents yo need to store,  dembedding techique as object)
db 

# How Do Query on VectorDB (retrieve the results from query vectorstore db)
query = 'It will be all the easier for us to conduct ourselves as belligerents'
retrived_result = db.similarity_search(query) # دا بيقيت التشابه بين النص اللى احنا مدخليه وكل ال vector اللى عندنا
print(retrived_result)
```

## 🧠 Sentence Transformers on Hugging Face
<div dir='rtl'>

### ✅ إيه هي sentence-transformers؟
- **sentence-transformers** هي مكتبة في Python بتستخدم علشان تحوّل:
  - الجُمل (Sentences)
  - النصوص (Texts)
  - الصور (Images)
- لتمثيلات رقمية (Vectors) بتعبّر عن **المعنى** الحقيقي للكلام.
- بتُستخدم في:
  - البحث الدلالي (Semantic Search)
  - حساب التشابه (Similarity Matching)
  - التجميع (Clustering)
  - التصنيف (Classification)

### ✅ استخدامها في LangChain:
- LangChain بتوفر كلاس اسمه **`HuggingFaceEmbeddings`** بيستخدم موديلات **sentence-transformers**.
- كمان فيه اسم مختصر (alias) اسمه **`SentenceTransformerEmbeddings`** لنفس الوظيفة.

### ✅ طريقة العمل:
1. تدخل جملة زي: `"هذا نموذج ذكاء اصطناعي."`
2. الجملة بتدخل على الموديل (sentence-transformer).
3. الموديل بيطلع تمثيل رقمي (vector) مثلاً زي: `[0.12, -0.98, 0.45, ..., 0.34]` بيمثّل معنى الجملة.

### ✅ ليه نستخدمها؟
| الخاصية | الفائدة |
|----------|---------|
| 🧠 الفهم الدلالي | بيعبر عن معنى الجملة مش مجرد كلمات. |
| 🔍 التشابه والبحث | تقدر تلاقي نصوص شبه بعضها بسهولة. |
| 🚀 دمج سهل | LangChain مجهزة الموضوع بـ HuggingFaceEmbeddings. |

```python
load_dotenv()
os.environ['HF_TOKEN'] = os.getenv('HF_TOKEN')
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
# بنجيب اسم الموديل من huggingface
# link model => https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
text = "this is a test documents"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
# we can use multiple documents or texts
doc_result = embeddings.embed_documents([text, 'this is not a test document'])
doc_result[0] # first embedding
```
